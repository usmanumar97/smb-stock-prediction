### Correlation Analysis

```python
df_corr = df.drop(['Date', 'S-Day Trend', 'SP-Day Trend', 'BI-Day Trend', 'PBI-Day Trend', 'I-Day Trend', 'PI-Day Trend'], axis=1)
```

```python
df_corr.dtypes
```

### Converting the objects to float

```python
for col in df_corr.columns:
    if df_corr[col].dtype == 'object':
        df_corr[col] = df_corr[col].str.replace(',', '').replace('-', np.nan).astype(float)
    
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(data=df_corr.corr(), annot=False, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Correlation Heatmap", fontsize=16)
plt.show()
```

```python
df_corr.corr()
```

```python
columns_to_drop = [
    'S-Open', 'S-High', 'S-Low',
    'SP-Open', 'SP- High', 'SP- Low',
    'BI-Open', 'BI-High', 'BI-Low',
    'PBI-Open', 'PBI-High', 'PBI-Low',
    'PI-Open', 'PI-High', 'PI-Low',
    'S-Avg Shares per Trade', 'SP- Avg Shares per Trade', 'BI-Avg Shares per Trade',
    'PBI-Avg Shares per Trade', 'PI-Avg Shares per Trade',
    'S-Vol Traded', 'SP- Vol Traded', 'BI-Volume Traded',
    'PBI-Volume Traded', 'PI-Volume Traded'
]
```

```python
df_corr.drop(columns=columns_to_drop, axis=1, inplace=True)
```

#### Updating the pre processing details in the original dataframe

- Dropping Columns
- Changing Data Type

```python
df.drop(columns=columns_to_drop, axis=1, inplace=True)

df.head()
```

```python
ignore_columns = ['Date', 'S-Day Trend', 'SP-Change', 'SP-Day Trend', 'BI-Day Trend', 'PBI-Day Trend', 'I-Day Trend', 'PI-Day Trend']

for col in df.columns:
    if col in ignore_columns:
        continue
    if df[col].dtype == 'object':
        df[col] = df[col].str.replace(',', '').replace('-', np.nan).astype(float)
```
