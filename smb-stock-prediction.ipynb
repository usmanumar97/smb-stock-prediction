### Pre processing the cateogorical data

```python
df_cat = df[['S-Day Trend', 'SP-Day Trend', 'BI-Day Trend', 'PBI-Day Trend', 'I-Day Trend', 'PI-Day Trend']]
```

#### Mapping the values

```python
trend_mapping = {}
for trend in df_cat['S-Day Trend'].unique():
    direction, magnitude = trend.split('-') if '-' in trend else(trend,0)
    if direction == 'Up':
        trend_mapping[trend] = int(magnitude)
    elif direction == 'Down':
        trend_mapping[trend] = -int(magnitude)
    else:
        trend_mapping[trend] = 0
```

```python
# Apply the mapping
df['S-Day Trend'] = df['S-Day Trend'].map(trend_mapping)
df['SP-Day Trend'] = df['SP-Day Trend'].map(trend_mapping)
df['BI-Day Trend'] = df['BI-Day Trend'].map(trend_mapping)
df['PBI-Day Trend'] = df['PBI-Day Trend'].map(trend_mapping)
df['I-Day Trend'] = df['I-Day Trend'].map(trend_mapping)
df['PI-Day Trend'] = df['PI-Day Trend'].map(trend_mapping)
```

#### Addressing the null values int the dataset

```python
df.loc[pd.isna(df['I-Day Trend']) & (df['PBI-Day Trend'] == 3), 'I-Day Trend'] = 3
df.loc[pd.isna(df['I-Day Trend']) & (df['PBI-Day Trend'] == -1), 'I-Day Trend'] = -1
df.loc[pd.isna(df['PI-Day Trend']) & (df['PBI-Day Trend'] == 2), 'PI-Day Trend'] = 2
df.loc[pd.isna(df['PI-Day Trend']) & (df['PBI-Day Trend'] == 3), 'PI-Day Trend'] = 3
```

### Feature Engineering

```python
df['10_day_EMA'] = df['S-Close'].ewm(span=10, adjust=False).mean()
df['30_day_EMA'] = df['S-Close'].ewm(span=30, adjust=False).mean()

# Create a crossover feature
df['EMA_Crossover'] = df['10_day_EMA'] - df['30_day_EMA']

df.head()
```

# Model Creation

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
import matplotlib.pyplot as plt

# Ensure '10-day EMA' exists in the DataFrame
if '10-day EMA' not in df.columns:
    df['10-day EMA'] = df['S-Close'].ewm(span=10, adjust=False).mean()

# Use EMA as the main feature for prediction
data = df[['S-Close', '10-day EMA']].copy()

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(data)

# Create rolling windows for input (X) and output (y)
def create_sequences(data, window_size):
    X, y = [], []
    for i in range(window_size, len(data)):
        X.append(data[i-window_size:i, :])  # Last 'window_size' days
        y.append(data[i, 0])  # Next day's 'S-Close' price
    return np.array(X), np.array(y)

window_size = 10
X, y = create_sequences(data_scaled, window_size)

# Split into training and testing sets
train_size = int(0.8 * len(X))
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]


# Step 2: Build the LSTM Model
model = Sequential([
    LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
    LSTM(units=50),
    Dense(units=1)  # Predicting the next day's price
])

model.compile(optimizer='adam', loss='mean_squared_error')

# Step 3: Train the Model
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test), verbose=1)

# Step 4: Predictions
predicted_prices = model.predict(X_test)
predicted_prices = scaler.inverse_transform(np.concatenate((predicted_prices, np.zeros((predicted_prices.shape[0], 1))), axis=1))[:, 0]

# Reverse scaling for actual prices
actual_prices = scaler.inverse_transform(np.concatenate((y_test.reshape(-1, 1), np.zeros((y_test.shape[0], 1))), axis=1))[:, 0]

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(actual_prices, label='Actual Prices', marker='o')
plt.plot(predicted_prices, label='Predicted Prices', marker='x')
plt.title('Actual vs Predicted Stock Prices (LSTM)')
plt.xlabel('Time (Test Set Index)')
plt.ylabel('Price')
plt.legend()
plt.show()

# Evaluate the model
mse = np.mean((actual_prices - predicted_prices)**2)
print(f"Mean Squared Error: {mse:.2f}")
```
